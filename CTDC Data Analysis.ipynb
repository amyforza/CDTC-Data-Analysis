```python
from collections import Counter

mc = df[[cn for cn in df.columns if 'meansOfControl' in cn]].copy()

mc.columns = [cn[14:] for cn in mc.columns]

tag = 'Concatenated'
means = mc[~mc[tag].isin(['-99'])][tag].tolist()

bag = list()
for m in means:
    bag += m.split(';')
    
count = Counter(bag).most_common(50)

values = [c[1] for c in count]
labels = [c[0] for c in count]

## Create a matplotlib figure
fig = plt.figure(figsize=(12,4), dpi = 600)
ax = fig.add_subplot(111)

plot_df = pd.DataFrame(count, columns = ['Means','Count'])
plot_df.sort_values(by = ['Means'], ascending = True, inplace = True)

plot_df.set_index('Means').plot(ax = ax,
                                kind = 'bar', 
#                                color = '#345454',
                                legend = False)

## Adding Title, X-axis, and Y-axis labels
plt.title('Means of Control', fontweight = 'bold', fontsize = 18)
plt.xlabel('Means of Control', fontweight = 'normal', fontsize = 14)
plt.ylabel('Case Count', fontweight = 'normal', fontsize = 14)

## Save figure
#plt.savefig('Means of Control.png', dpi = 600, transparent = True, bbox_inches = 'tight')
```
