```python
df['Count'] = 0
countries = df.groupby('Country-of-Origin').count()[['Count']]
countries.reset_index(inplace = True)
countries.sort_values(by = ['Count'], ascending = False, inplace = True)

total = countries['Count'].sum()

countries['%'] = countries['Count'].apply(lambda t: (t/total) * 100)

countries.iloc[[0,2,3,4,5,6]]
```

```python
## -----------------------------------------------------------------------------
def format_plot(ax, font_size, color):
    
    ax.tick_params(direction='out', length = 4, width = 1, colors = color,
                   labelsize = font_size, labelcolor = color)

    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    
    return

countries = [c for c in Counter(df['Country-of-Origin'], ascending = True, inplace = True).most_common(11) if 'Unknown' not in c]
plot_df = pd.DataFrame(countries,columns = ['Country','Count']).sort_values(by = 'Count', ascending = True)
font_size = 10.0
background = 'black'

fig = plt.figure(figsize=(12,4), dpi = 600)
ax = fig.add_subplot(111)

plot_df.set_index('Country').plot(ax = ax, kind = 'barh', legend = False)

plt.title('Top 10 Human Trafficking by Country', fontweight = 'bold', fontsize = 18, color = background)
plt.xlabel('Number of Cases', fontweight = 'normal', fontsize = 12, color = background)
plt.ylabel('Country', fontweight = 'normal', fontsize = 12, color = background)

format_plot(ax, font_size, background)

plt.savefig('trafficking_countries.png', transparent = True, dpi = 600, bbox_inches = 'tight')
```
